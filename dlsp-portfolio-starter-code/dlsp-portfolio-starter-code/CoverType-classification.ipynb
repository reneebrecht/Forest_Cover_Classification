```python
class_distribution = df['class'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
class_distribution.plot(kind='bar')
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Split the data into features (X) and target labels (y)
X = df.drop("class", axis=1)  # Features: Remove the target column
y = df["class"]  # Target labels
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)  # Adjust test_size and random_state as needed
```
